```python
# Suppress warnings
import warnings
warnings.simplefilter('ignore')
warnings.filterwarnings('ignore')
from pprint import pprint

import earthaccess 
import xarray as xr
```

```python
auth = earthaccess.login()
# are we authenticated?
if not auth.authenticated:
    # ask for credentials and persist them in a .netrc file
    auth.login(strategy="interactive", persist=True)
```

```python
short_name = 'MUR-JPL-L4-GLOB-v4.1'
version = "4.1"
date_start = "2021-06-01"
date_end = "2021-06-30"
date_range = (date_start, date_end)
# min lon, min lat, max lon, max lat
bbox = (10, 10, 20, 20)  

results = earthaccess.search_data(
    short_name = short_name,
    version = version,
    cloud_hosted = True,
    temporal = date_range,
    bounding_box = bbox,
)
len(results)
```

```python
fileset = earthaccess.open(results[0:1])
ds = xr.open_dataset(fileset[0])
ds
```

```python
for v in ds.variables:
    print(v)
```

```python
ds.variables['analysed_sst'].attrs
```

```python
x_lon_coords = [-26, -15]
y_lat_coords = [10, 18]

ds_subset = ds.sel(time=date_start, lat=slice(min(y_lat_coords), max(y_lat_coords)), lon=slice(min(x_lon_coords), max(x_lon_coords))) 
ds_subset
```

```python
ds_subset['analysed_sst'].plot(figsize=(10,6), x='lon', y='lat');
```

## multi-file test

```python
fileset = earthaccess.open(results[1:31])

# Load data; takes awhile
ds = xr.open_mfdataset(fileset, chunks = {})

# Get SST data 
ds_subset = ds['analysed_sst']
```

```python
ds_subset
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter

plt.figure(figsize=(14, 10))

# Label axes of a Plate Carree projection with a central longitude of 180:
ax1 = plt.subplot(211, projection=ccrs.PlateCarree(central_longitude=180))

# Use the lon and lat ranges to set the extent of the map
# the 120, 260 lon range will show the whole Pacific
# the 15, 55 lat range with capture the range of the data
#ax1.set_extent([260, 350, 15, 55], ccrs.PlateCarree())
ax1.set_extent([260, 350, -25, 25], ccrs.PlateCarree())

# set the tick marks to be slightly inside the map extents

# add feature to the map
ax1.add_feature(cfeature.LAND, facecolor='0.6')
ax1.coastlines()

# format the lat and lon axis labels
lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()
ax1.xaxis.set_major_formatter(lon_formatter)
ax1.yaxis.set_major_formatter(lat_formatter)

sst_cm = ds_subset[0].plot.pcolormesh(ax=ax1, transform=ccrs.PlateCarree(), cmap='jet')
plt.title('Satellite Data Before Masking')
```
